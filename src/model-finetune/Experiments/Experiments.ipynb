```python
f1 = open('pol-merged.txt')
f2 = open('pol_clean.txt', 'w+')
s = f1.readline()

while True:
    s = f1.readline()
    if not s:
        break
    if((s[0:2] == ">>") or (s[0:2] == "--") or (s == "\n") or (s[0:4] == "http")):
        continue
    else:
        s = s.replace(">", "")
        if(s != "\n"):
            if s[-2].isalpha():
                s = s[0:-1] + '.' + '\n'
            f2.write(s)
    
    
f1.close()
f2.close()
```

```python
f1 = open('kjv.txt')
f2 = open('bible_clean.txt', 'w+')
s = f1.readline()

while True:
    s = f1.readline()
    if not s:
        break
    for i in range(len(s)):
        if s[i] == ' ':
            break

    f2.write(s[i+1:])
    
f1.close()
f2.close()
```
